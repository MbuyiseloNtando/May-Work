```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('https://raw.githubusercontent.com/MbuyiseloNtando/May-Work/main/y.csv')
```

```python
df
```

Regression on known values to predict the unknown

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
df.head()

testdf = df[df['n'].isnull()==True]       #year of Nan values as test
traindf = df[df['n'].isnull()==False]     # years of Not Nan values

y = traindf['n']     #Not Nan values
traindf.drop("n",axis=1,inplace=True)

lr.fit(traindf,y)
testdf.drop("n",axis=1,inplace=True)
pred = lr.predict(testdf)
testdf['n']= pred
```

```python
testdf
```

```python
series = pd.Series(pred)
series
```

```python
df.n.fillna(series, inplace=True)
```
